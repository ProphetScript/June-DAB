```python
##Replacing missing values (numerical columns)
inn_df['reorder_level'] = inn_df['reorder_level'].fillna(inn_df['reorder_level'].mean())
sales_df['store_id'] = sales_df['store_id'].fillna(sales_df['store_id'].mean())
stores_df['size_sqft'] = stores_df['size_sqft'].fillna(stores_df['size_sqft'].mean())
product_df['weight'] = product_df['weight'].fillna(product_df['weight'].mean())
```

```python
##Replacing missing values (Categorical columns)
inn_df['last_restock_date'] = inn_df['last_restock_date'].fillna(inn_df['last_restock_date'].mode())
sales_df['date'] = sales_df['date'].fillna(sales_df['date'].mode()) 
sales_df['payment_method'] = sales_df['payment_method'].fillna(sales_df['payment_method'].mode())
stores_df['opening_date'] = stores_df['opening_date'].fillna(stores_df['opening_date'].mode())
customers_df['email'] = customers_df['email'].fillna(customers_df['email'].mode())
customers_df['phone'] = customers_df['phone'].fillna(customers_df['phone'].mode())
customers_df['zip_code'] = customers_df['zip_code'].fillna(customers_df['zip_code'].mode())
customers_df['registration_date'] = customers_df['registration_date'].fillna(customers_df['registration_date'].mode())
```

```python
##Deleting duplicated entries in customers and product data frames
customers_df.drop_duplicates()
```
